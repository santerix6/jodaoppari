Tätä mallia sitten hyödynnettiin erilaisiin tarkoituksiin. Lähinnä siihen, että annettiin mallin arvioida kuka ministeri on sanonut lauseen, joka mallille syötetään. 

```python
#!pip install nltk
#!pip install bs4
import urllib.request
import nltk
#nltk.download('stopwords') 
from nltk.corpus import stopwords 
from bs4 import BeautifulSoup

response =  urllib.request.urlopen('https://en.wikipedia.org/wiki/Python_(programming_language)')
html = response.read()
soup = BeautifulSoup(html)
text = soup.get_text(strip = True)
tokens = [t for t in text.split()]
sr = stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in sr:       
        clean_tokens.remove(token)
print(len(clean_tokens), len(tokens))
freq = nltk.FreqDist(clean_tokens)
freq.plot(20, cumulative=False)
```

Yllä on tämän viikon koodiesimerkki. Siinä haetaan Python koodikielen wikipedia sivu ja tarkastellaan siinä käytettyjä sanoja. Lisäksi haetaan nltk kirjaston stopwords lista ja tarkastellaan kuinka monta sanaa lähtee kun poistetaan stopwordit. Nähdään, että sanoja lähtee noin 2000.
