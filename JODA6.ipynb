Luento oli jälleen melko käytännön läheinen ja melkein koko luento käytettiin koodin läpikäymiseen mikä ei mielestäni ole mitenkään huono asia. Ensimmäisenä käytiin läpi ryvästämistä, jossa datasewtin pisteet jaetaan ryppäisiin siten, että kukin rypäs sijaitsee mahdollisimman pienellä alueella. Ryvästämistä tehdään usein KMeans algoritmeilla. Omin sanoin sanoisin, että KMeans algoritmien idea on löytää keskuspisteet syötteen pisteiden perusteella ja K arvon määrällä ja sitten pisteet jakautuvat keskuspisteiden mukaan eri klustereihin. Luennolla käytiin läpi elokuviin liittyvää dataa, joka löytyi Kaggle palvelusta, josta löytyy paljon erilaisia mielenkiintoisia datasettejä. Datasetti, jota luennolla käytettiin sisälsi 24 eri saraketta, joista monet olivat kategorisoivia. Näitä ei voi luennon mukaan käyttää KMeans algoritmissä. Kategorista dataa voi luennon mukaan ryvästää KModes-ryvästämisellä. Tästä oli linkattu artikkeli, jota silmäilin, mutta artikkeli oli melko raskasta luettavaa, joten siitä ei jäänyt hirveästi käteen. Lähdin googlaamaan esimerkkejä KModes algoritmeistä ja löysinkin mielenkiintoisia artikkeleja. Vaikuttaa siltä, että KModessilla saa tehtyä mielenkiintoisia analyyseja. Luennon KMEeans algoritmissä otettiin käyttöön vaa arvot, joissa on numeeresia arvoja sisältävät sarakkeet. Dataa lisäksi siistittiin, siten että nollaarvoja sisältävät arvot poistettiin. KMeans algoritmissa pitää valita klustereiden määrä. Se ei aina ole helppoa, mutta yksi lähestymistapa on kyynerpääperiaate. Tätä periaatetta olen käyttänyt aikaisemmin muutamassa projektissa. Luennolla käytiin läpi myös luonnollisen kielen analyysia ja aihemallinusta, joka oli hyvää jatkoa viime viikon luennolle, jossa käsiteltiin NLP:tä enemmän. Luennolla keskustelua herätti se, että oliko luennolla käytetty aihemallinnus hyödyllistä vai ei. Mielestäni aiheisiin liittyvät sanat olivat osittain toisiinsa liittyviä ja patterneja näkyi, mutta yleisesti aihemallinnuksesta jäi hieman epäselvä ja epävarma kuva. Tätä voisi lähteä tutkimaan lisää itse, jos haluaisin lisää tietoa ja ymmärrystä aiheesta. 

```python

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Yllä olevassa koodiesimerkissä luodaan make_blobs funktiolla pisteitä, jotka muodostavat erilaisia clustereita randomilla. Sitten ajetaan KMeans algorimti datasetille k arvoilla 1-11. Tästä saadaan piirrettyä kynerpäägraphi, josta nähdään, että 4 on clustereiden oikea määrä. Tästä saisi sitten vielä lisää tietoa mitä ikina halutaankin KMeansilla tehdä.
